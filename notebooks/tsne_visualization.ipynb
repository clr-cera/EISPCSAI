```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
tsne_features = np.load("../features/tsne_features.npy")
dfy = pd.read_csv("../sentiment-dataset/annotations.csv", sep=";")
imgname = dfy.iloc[:, 1].values
imgname = list(map(lambda x: "../sentiment-dataset/images/" + x, imgname))
colors = dfy.iloc[:, 2].values


```

```python
print(imgname)
```

```python
tsne_df = pd.DataFrame(data=tsne_features, columns=['x', 'y'])
tsne_df['image'] = imgname
tsne_df['color'] = colors
```

```python
alt.Chart(tsne_df).mark_circle(size=30).encode(
    x="x",
    y="y",
    color=alt.Color('color:N', scale=alt.Scale(scheme='lightgreyred')),
    tooltip=['image']
).properties(
    width=2400,
    height=1800
).interactive().configure_view(strokeWidth=0).save("tsne_visualization.html")
```

```python
chart = alt.Chart(tsne_df[::2]).mark_image(
    width=12*4,
    height=12*4).encode(
    x="x",
    y="y",
    url="image",
    ).properties(
    width=800*4,
    height=600*4
)
```
