```python
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
pca_features = np.load("../features/tsne_features.npy")
dfy = pd.read_csv("../sentiment-dataset/annotations.csv", sep=";")
imgname = dfy.iloc[:, 1].values
colors = dfy.iloc[:, 2].values
```

```python
print(imgname)
```

```python
plotly_fig = px.scatter(
    x=pca_features[:, 0],
    y=pca_features[:, 1],
    color=colors,
    title='TSNE Visualization of Features',
    hover_data={'img_name': imgname},
)

plotly_fig.show()
```
